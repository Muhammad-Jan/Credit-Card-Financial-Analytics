# Cleaning (credit_card -- Dataset)

```python
import pandas as pd
import numpy as np
```

### Load Credit Card Dataset

```python
df = pd.read_csv("credit_card.csv")
df.head(3)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

### Clean Card_Category Column
##### - Remove extra spaces
##### - Standardize text format
##### - Correct inconsistent values
##### - Fill missing values

```python
df['Card_Category'].value_counts(dropna = False)
```

```python
df['Card_Category'] = df['Card_Category'].str.strip().str.title()
df['Card_Category'] = df['Card_Category'].apply(lambda x:'Blue' if x == 'Blu' else x)
df['Card_Category'] = df['Card_Category'].bfill()
df['Card_Category'].value_counts(dropna = False)
```

```python
df.head(5)
```

### Handle Missing Values in Customer Acquisition Cost

```python
df['Customer_Acq_Cost'] = df['Customer_Acq_Cost'].bfill()
```

```python
df.isnull().sum()
```

### Clean Credit Limit Column
##### - Convert negative values to positive
##### - Fill missing values

```python
df['Credit_Limit'] = df['Credit_Limit'].abs()
df['Credit_Limit'] = df['Credit_Limit'].ffill()
```

### Handle Missing Values in Avg_Utilization_Ratio

```python
df['Avg_Utilization_Ratio'].value_counts()
```

```python
df['Avg_Utilization_Ratio'] = df['Avg_Utilization_Ratio'].fillna(method = 'bfill')
```

```python
df.isnull().sum()
```

### Clean Use Chip Column

##### - Standardize text
##### - Correct spelling
##### - Fill missing values

```python
df['Use Chip'].value_counts()
```

```python
df['Use Chip'] = df['Use Chip'].str.strip().str.title()
df['Use Chip'].value_counts()
```

```python
df['Use Chip'] = df['Use Chip'].apply(lambda x:'Swipe' if x == 'Swip' else x)
df['Use Chip'] = df['Use Chip'].ffill()
```

```python
df.isnull().sum()
```

### Clean Expense Type Column
##### - Standardize text
##### - Correct spelling
##### - Fill missing values

```python
df['Exp Type'].value_counts()
```

```python
df['Exp Type'] = df['Exp Type'].str.strip().str.title()
df['Exp Type'].value_counts()
```

```python
df['Exp Type'] = df['Exp Type'].apply(lambda x: 'Entertainment' if x == 'Entertainmnt' else x)
df['Exp Type'].value_counts(dropna = False)
```

```python
df['Exp Type'] = df['Exp Type'].fillna(method = 'ffill')
```

```python
df.isnull().sum()
```

### Remove Remaining Missing Values

```python
df = df.dropna()
```

### Export Clean Credit Card Dataset

```python
df.to_csv("cleaned_credit_card.csv",index = False)
```
